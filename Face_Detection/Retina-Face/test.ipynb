# Import Libraries

```python
from retinaface import RetinaFace
import glob
import cv2
import matplotlib.pyplot as plt
```

```python
file_type=['JPG','JPEG','PNG','JFIF']
images=[]

for format in file_type:
    for path in glob.glob(f"images/*.{format}"):
        images.append(cv2.imread(path))
```

```python
for index in range(len(images)):
    detection = RetinaFace.detect_faces(images[index])
    for key in detection.keys():
        facial_area = detection["facial_area"]
    
        cv2.rectangle(images[index], (facial_area[2],facial_area[3]),(facial_area[0],facial_area[1]),(0,255,0),1)
    
    plt.show(images[index][:,:,::-1])
    print("---------------------------------")
```
