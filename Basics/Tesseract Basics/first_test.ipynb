```python
from PIL import Image
import pytesseract
from pytesseract import Output
import argparse
import cv2
import os
import matplotlib.pyplot as plt
```

```python
pytesseract.pytesseract.tesseract_cmd = r'C://Program Files//Tesseract-OCR//tesseract.exe'

os.environ["TESSDATA_PREFIX"] =  "C://Program Files//Tesseract-OCR//tessdata"
custom_config = r'-c preserve_interword_spaces=1'
```

```python
image = cv2.imread("Capture.PNG")
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
```

```python
plt.imshow(image)
```

```python
plt.imshow(gray)
```

### With Thresh

```python
gray_thresh = cv2.threshold(gray, 0, 255, cv2.THRESH_BINARY | cv2.THRESH_OTSU)[1]
gray_thresh = Image.fromarray(gray_thresh)
plt.imshow(gray_thresh);
```

### With Blur

```python
gray_blur = cv2.medianBlur(gray, 1)
gray_blur = Image.fromarray(gray_blur)
plt.imshow(gray_blur);
```

### Image to text

```python
text_thresh = pytesseract.image_to_string(gray_thresh,lang = 'eng', config = custom_config)
```

```python
# text_blur   = pytesseract.image_to_data(gray_blur,lang = 'eng', config = custom_config,output_type=Output.DICT)['text']
text_blur = pytesseract.image_to_string(gray_thresh,lang = 'eng', config = custom_config)
```
