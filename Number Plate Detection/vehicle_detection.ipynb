```python
import torch
import matplotlib.pyplot as plt
from ultralytics import YOLO
```

```python
model_v8 = YOLO('yolov8n.pt')
model_v5 = YOLO('yolov5s.pt')
```

### Testing on Images

```python
import glob
import cv2
```

### importing images

```python
file_type = ['JPG','JPEG','PNG']
images = [] 

for format in file_type:
    for path in glob.glob(f"images/*.{format}"):
        img = cv2.imread(path)
        img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
        images.append(img)
```

```python
cv2.imshow('Image',img)
cv2.waitKey(0)
```

```python
def plot_img(img):
    plt.axis('off')
    plt.imshow(img)
    plt.show()
```

### Predict

```python
for i in range(len(images)):
    predict_image = model_v8(images[i])
    res_plot = predict_image[0].plot()
    boxes = predict_image[0].boxes.data
    print(boxes)
    # print(res_plot.xyxy[0])
    plot_img(res_plot)
    # cv2.waitKey(0)
    # print(res_plot)
```

```python
for i in range(len(images)):
    predict_image = model_v5(images[i])
    res_plot = predict_image[0].plot()
    plot_img(res_plot)
    # cv2.waitKey(0)
    # print(res_plot)
```

```python
cropped_image = cv2.cvtColor(cropped_image, cv2.COLOR_RGB2BGR)
```

```python
import cv2
import imutils
import pytesseract
import os
```

```python
pytesseract.pytesseract.tesseract_cmd = r'C://Program Files//Tesseract-OCR//tesseract.exe'

os.environ["TESSDATA_PREFIX"] =  "C://Program Files//Tesseract-OCR//tessdata"
custom_config = r'-c preserve_interword_spaces=1'
```

```python
gray_image = cv2.cvtColor(cropped_image, cv2.COLOR_RGB2GRAY)
cv2.imshow("Gray Image",gray_image)
cv2.waitKey(0)
```

```python
gray_image = cv2.bilateralFilter(gray_image, 11, 17, 17)
cv2.imshow("Smoothened Image", gray_image)
cv2.waitKey(0)
```

```python
edged = cv2.Canny(gray_image, 50, 200)
cv2.imshow("Edged",edged)
cv2.waitKey(0)
```

```python
cnts,new = cv2.findContours(edged.copy(), cv2.RETR_LIST, cv2.CHAIN_APPROX_SIMPLE)
image1 = cropped_image.copy()
cv2.drawContours(image1, cnts, -1,(0,255,0),1)
cv2.imshow("Contours",image1)
cv2.waitKey(0)
```

```python
cnts = sorted(cnts, key = cv2.contourArea, reverse = True)[:30]
screenCnt = None
image2 = cropped_image.copy()
cv2.drawContours(image2, cnts, -1, (0,255,0),1)
plt.imshow("top 30 Contours", image2)
cv2.waitKey(0)
```

```python
i = 7
for c in cnts:
    perimeter = cv2.arcLength(c, True)
    approx = cv2.approxPolyDP(c, 0.018 * perimeter, True)
    if len(approx) == 4:
        screenCnt = approx
        # Cropping the rectangular part identified as license plate
        x,y,w,h = cv2.boundingRect(c)
        new_img = cropped_image[y:y+h, x:x+w]
        cv2.imwrite('./'+str(i)+'.png',new_img)
        i+=1
        break
```

```python
plate = pytesseract.image_to_string(cropped_image,lang = 'eng', config = custom_config)

print(f"Number plate is {plate}")
# cv2.waitKey(0)
# cv2.destroyAllWindows()
```
